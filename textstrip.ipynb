```python
import pandas as pd
rating_ensemble = pd.DataFrame()
```

```python
dratings_file = open('./dratings.txt', 'r')
dratings_content = dratings_file.readlines()
dratings_file.close()
```

```python
def try_float(x):
    try:
        float(x)
    except:
        if ')' in x or '-' in x:
            return 0
        else:
            return x
    else:
        return float(x)
```

```python
dratings= [ x.strip() for x in dratings_content ]
dratings = [ try_float(x) for x in dratings ]
dratings = list(filter( lambda x: type(x) == str, dratings ))
dratings[:11]
```

```python
rating_ensemble['dratings'] = dratings
```

```python
fifa_file = open('./fifaelo.txt', 'r')
fifa_content = fifa_file.readlines()
fifa_file.close()
```

```python
fifa = [ x.split('\t') for x in fifa_content ]
fifa[0]
```

```python
fifaelo = []
for x in fifa:
    try:
        fifaelo.append(x[1][3:])
    except:
        pass
fifaelo[:11]
```

```python
fifae = pd.DataFrame({ 'fifaelo': fifaelo })
```

```python
rating_ensemble = pd.concat([rating_ensemble, fifae], axis=1) 
```
