```python
neutral_filter = results['neutral'] == True
results[neutral_filter].shape
```

```python
results = results.drop(['city', 'country', 'tournament'], axis=1)
```

```python
results['draw'] = results.apply(lambda row: row.home_score == row.away_score, axis=1)
results['home_win'] = results.apply(lambda row: row.home_score > row.away_score, axis=1)
results['away_win'] = results.apply(lambda row: row.home_score < row.away_score, axis=1)
results['margin_win'] = results.apply(lambda row: row.home_score - row.away_score, axis=1)
```

```python
results[ results['date'].str.contains('1938') ].head()
```

```python
results = results[2005:]
results.shape
```

```python
results = results.reset_index(drop=True)
results.head()
```
